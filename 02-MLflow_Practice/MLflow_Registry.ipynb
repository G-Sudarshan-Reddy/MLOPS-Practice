```python
from mlflow.tracking import MlflowClient

client = MlflowClient(tracking_uri='sqlite:///mlflow.db')
```

```python
client.create_experiment(name="my-new-experiment-1")
```

```python
import mlflow
mlflow.set_tracking_uri('sqlite:///mlflow.db')
```

```python
from mlflow.entities import ViewType

runs = client.search_runs(
    experiment_ids='3',
    filter_string="",
    run_view_type=ViewType.ACTIVE_ONLY,
    max_results=5,
    order_by=["metric.rmse ASC"]
)
```

```python
for run in runs:
    print(f"run id: {run.info.run_id}, rmse: {run.data.metrics['rmse']:.4f}")
```

```python
run_id = 'b172c2f233284d14997b376131a80512'
model_uri = f"runs:/{run_id}/model"

mlflow.register_model(model_uri=model_uri, name="new-regressor-3")
```

```python
client.get_latest_versions(name = "new-regressor-3")
```

```python
client.transition_model_version_stage(
    name="new-regressor-3",
    version=4,
    stage="Staging",
    archive_existing_versions=False
)
```
